```python
def load_images_from_folder(folder):
    images = []
    for file_name in os.listdir(folder):
        img = cv.imread(os.path.join(folder, file_name))
        if img is not None:
            images.append(img)
    return images
```

```python
images = load_images_from_folder("C:\Vishnu\semester Files\Machine learning\project\Images\Eminem")
for img in range(len(images)):
    images[img] = cv.cvtColor(images[img],cv.COLOR_BGR2GRAY)
    plt.imshow(images[img],cmap='gray')
    plt.show()
```

```python
face_classifier = cv.CascadeClassifier('C:\Vishnu\semester Files\Machine learning\project\haarcascades\haarcascade_frontalface_alt.xml')
faces = []
for img in images:
    temp = face_classifier.detectMultiScale(img, 1.0485258, 6)
    print(temp)
    faces.append(temp)

import numpy as np
# faces = np.array(faces)
# print(faces)
```
